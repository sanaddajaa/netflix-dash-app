Task 1 - Plotly Interactive Charts

Import Libraries

```python
import pandas as pd
import plotly.express as px
```

Load and Clean the Dataset

```python
df = pd.read_csv("netflix_titles.csv")

df_clean = df.copy()
df_clean['director'] = df_clean['director'].fillna('Unknown')
df_clean['cast'] = df_clean['cast'].fillna('Unknown')
df_clean['country'] = df_clean['country'].fillna('Unknown')

df_clean['date_added'] = df_clean['date_added'].str.strip()
df_clean['date_added'] = pd.to_datetime(df_clean['date_added'], errors='coerce')

df_clean.dropna(subset=['date_added'], inplace=True)

df_clean['year_added'] = df_clean['date_added'].dt.year

df_clean.drop_duplicates(inplace=True)

df_clean.head()
```

Chart 1 - Titles Added Over Time (converted from Matplotlib line chart)

```python
year_counts = df_clean['year_added'].value_counts().sort_index().reset_index()
year_counts.columns = ['Year', 'Count']

fig1 = px.line(year_counts, x='Year', y='Count', markers=True,
               title='Titles Added to Netflix Over Time')

fig1.show()
```

Insight: Netflix content additions increased a lot over the years, especially after 2015

Chart 2 - Movies vs TV Shows (converted from Matplotlib bar chart)

```python
type_counts = df_clean['type'].value_counts().reset_index()
type_counts.columns = ['Type', 'Count']

fig2 = px.bar(type_counts, x='Type', y='Count', color='Type',
              title='Movies vs TV Shows on Netflix', text='Count')

fig2.show()
```

Insight: Movies are more common than TV Shows on Netflix

Chart 3 - Release Year vs Year Added (converted from Matplotlib scatter plot)

```python
fig3 = px.scatter(df_clean, x='release_year', y='year_added', color='type',
                  hover_data=['title', 'country'],
                  title='Release Year vs Year Added to Netflix',
                  opacity=0.5)

fig3.show()
```

Insight: Most content is added close to its release date, but some older content is also added

Chart 4 - Animated Chart: Top Countries per Year (multi-trace animated figure)

```python
# get top 5 countries without Unknown
top5 = [c for c in df_clean['country'].value_counts().index if c != 'Unknown'][:5]

df_anim = df_clean[df_clean['country'].isin(top5)]

anim_data = df_anim.groupby(['year_added', 'country', 'type']).size().reset_index(name='Count')
anim_data = anim_data[anim_data['year_added'] >= 2015]

fig4 = px.bar(anim_data, x='country', y='Count', color='type',
              animation_frame='year_added', barmode='group',
              title='Top 5 Countries: Movies vs TV Shows per Year')

fig4.show()
```

Insight: The US leads content production every year, followed by India and the UK
